```python
import matplotlib.pyplot as plt
from log_data import get_loss_iterations

iterations_adam, loss_values_adam = get_loss_iterations("D:/Code/study/adam6.txt")
iterations_adadelta, loss_values_adadelta = get_loss_iterations("D:/Code/study/adadelta6.txt")
iterations_asgd, loss_values_asgd = get_loss_iterations("D:/Code/study/asgd6.txt")
iterations_rmsprop, loss_values_rmsprop = get_loss_iterations("D:/Code/study/rmsprop6.txt")



# Plotting
f = plt.figure()
f.set_figwidth(15)
plt.plot(iterations_adam, loss_values_adam, label='Adam', color='r', linewidth=0.5)
plt.plot(iterations_adadelta, loss_values_adadelta, label='Adadelta', color='g', linewidth=0.5)
plt.plot(iterations_asgd, loss_values_asgd, label='ASGD', color='b', linewidth=0.5)
plt.plot(iterations_rmsprop, loss_values_rmsprop, label='RMSprop', color='black', linewidth=0.5)


plt.title('Loss Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.xticks(range(0,5100,200))
plt.xlim((0, 5000))
plt.legend()
plt.grid()
plt.savefig('6.png', dpi=300)
plt.show()
```
