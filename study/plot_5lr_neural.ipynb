```python
import matplotlib.pyplot as plt
from log_data import get_loss_iterations

main_path = "D:/Code/study/logs/5/"
iterations_adam, loss_values_adam = get_loss_iterations(main_path + "adam-crop64-depth6")
iterations_adadelta_64_6, loss_values_adadelta_64_6 = get_loss_iterations(main_path + "adadelta-crop64-depth6")
iterations_adadelta_64_10, loss_values_asgd_64_10 = get_loss_iterations(main_path + "adadelta-crop64-depth10")
iterations_adadelta_16_6, loss_values_adadelta_16_6 = get_loss_iterations(main_path + "adadelta-crop16-depth6")
iterations_adadelta_64_10_2, loss_values_adadelta_64_10_2 = get_loss_iterations(main_path + "adadelta-crop64-depth10_2")




# Plotting
f = plt.figure()
f.set_figwidth(15)
plt.plot(iterations_adam, loss_values_adam, label='adam-crop64-depth6', color='r', linewidth=0.5)
plt.plot(iterations_adadelta_64_6, loss_values_adadelta_64_6, label='adadelta-crop64-depth6', color='g', linewidth=0.5)
plt.plot(iterations_adadelta_64_10, loss_values_asgd_64_10, label='adadelta-crop64-depth10', color='b', linewidth=0.5)
plt.plot(iterations_adadelta_16_6, loss_values_adadelta_16_6, label='adadelta-crop16-depth6', color='black', linewidth=0.5)
plt.plot(iterations_adadelta_64_10_2, loss_values_adadelta_64_10_2, label='adadelta-crop64-depth10_2', color='pink', linewidth=0.5)



plt.title('Loss Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.xticks(range(0,5100,200))
plt.xlim((0, 5000))
plt.legend()
plt.grid()
plt.savefig('5.png', dpi=300)
plt.show()
```
